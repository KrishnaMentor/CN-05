## Input()
- Used to accept user input from the *keyboard* as a **string**, which can be further type cast as per requirement..
- `user_name = input("Enter the Prompt....")`

```python
user_name = input("Enter the UserName: ")
print(user_name)
```

```python
user_name = input("Enter the UserName: ")
print(user_name)
```

```python
# Type Cast
val = input("Enter the Value: ") # int('11') -> 11
print(val)
print(type(val))
```

```python
val = int(input("Enter the Value: ")) # int('11') -> 11
print(val)
print(type(val))
```

```python
print(val * 11) # val = 11 * 11 = 121
```

```python
print(val) # 11
```

```python
val = val * 11
# val *= 11
print(val)
```

```python
x = int(input("Enter the value between 1 to 100 :"))
x = x + 10
print(x)
```

```python
x = int(input("Enter the value between 1 to 100 :"))
if x > 100:
    print("Invalid Input, Please Enter the Val Between 1-100")
else:
    x = x + 10
    print(x)
```

```python
x = int(input("Enter the value between 1 to 100 :"))
if x > 100:
    print("Invalid Input, Please Enter the Val Between 1-100")
else:
    x = x + 10
    print(x)
```

```python
# String Pattern 
user_name = input("Enter the UserName: ")
email_id = input("Enter the Email Address: ")
print("Hey " + user_name + " Welcome to the Python Course!, You are being registered with us " + email_id + " ." )
```

```python
# String Pattern f-string
user_name = input("Enter the UserName: ")
email_id = input("Enter the Email Address: ")
print(f"Hey {user_name}, Welcome to the Python Course!, You are being registered with us {email_id} ✨" )
```
