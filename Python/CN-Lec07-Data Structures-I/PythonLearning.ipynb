# Comparison Operators (Return Boolean Result)
- '==' Equal To
- '!=' Not Equal To
- '>' Greater Than
- '>=' Greater Than or Equal To
- '<' Less Than
- '<=' Less Than or Equal To

```python
x = 11
y = 7
z = 15
print(x==y) # False
print(x!=y) # True
print(z!=y) # True
print(z!=15) # False
print(x>y) # True
print(x>=y) # True
print(z<=y) # False
print(z<x) # False
print(x!=2) # True
print(x>=3) # True
```

# Logical Operators: ('not' > 'and' > 'or')
- 'and' : Returns True, if both the conditions are True else False
- 'or' : Returns False, if both the conditions are False, else True
- 'not' : Returns the Opposite.

```python
x = 11
y = 7
z = 15
print((x==y) and (z!=y) or (y>x)) # (F and T or F) # (F or F) # F
print((x!=y) or (z!=y) or (y>x)) # (T or T or F) # (T or F) # T
print((x>=y) and (z==y) and (z>x)) # (T and F or T) # (F and T) # F
print(not(z!=15)) #Not(F) # T
print(not((x<=y) and (z>=y) or (y!=x))) # not(F and T or T) # not(F or T) # not(T) # ~T = F
```

```python
print(not(x<=y) and (z>=y) or (y!=x)) # not(F) and T or T  # T and T or T # True
```

```python
print(not(x<=y) or (z<=y) and (y!=x)) # not(F) or (F) and T 
# T or F and T # T or F # True
```

# Bitwise Operators: 
- 'AND' = '&' - 'Both bits with value 1 return 1 else 0'
- 'OR' = '|' - 'Any bit with value 1 return 1 else 0'
- 'XOR' = '^' - 'Alternative Bits returns 1 else same bit returns 0'
- 'NOT' = '~' - '2's complement to check the sign and evaluate'
- 'Left Shift' = '<<' - 'Shifts the bit to the left'
- 'Right Shift' = '>>' - 'Shifts the bit to the right'

```python
# And Bitwise
x = 10
y = 6
print(x & y) # 2
```

```python
# OR Bitwise
x = 10
y = 6
print(x | y) # 14
```

```python
# OR Bitwise
x = 19
y = 41
print(x | y) # 59
```

```python
# XOR Bitwise
x = 10
y = 6
print(x ^ y) # 12
```

```python
# NOT Bitwise (~)
X = 10
print(~X) # -(X+1) # -11
```

```python
# NOT Bitwise (~)
X = 7
print(~X) # -(X+1) # -8
```

```python
# NOT Bitwise (~)
X = -17
print(~X) # -(X+1) # -(-17+1) = -(-16) = 16
```

```python
# Left Shift (<<)
X = 10
print(X<<2) # 40
```

```python
# Right Shift (>>)
X = 10
print(X>>2) # 2
```

```python
# Right Shift (>>)
X = 10
print(X>>3) # 10 // 2^3 = 10 // 8 = 1
```

# Membership Operators : 
- It return the Boolean result.
- **'in'** operator -> `Return True if the value exist in a sequence` 
- **'not in'** operator -> `Return True if the value doesn't exist in a sequence`

```python
# Python is case-sensitive
print('hello' in 'hello World') # True
print('hello' in 'Hello World') # False
print('hello' not in 'Hello World') # True
```

```python
print('I' in 'India') # True
print('I' not in 'America') # True 
print('I' not in 'AmerIca') # False 
```

```python
print('mon' in ['Mon','Tue','Wed','Thurs','Fri']) # False
print('Sat' not in ['Mon','Tue','Wed','Thurs','Fri']) # True
print('Fri' not in ['Mon','Tue','Wed','Thurs','Fri']) # False
```

```python
# Dictionary {Key : Values}
print('mon' in {'mon':'Mon', 'tues':'Tues'}) # True
print('Mon' in {'mon':'Mon', 'tues':'Tues'}) # False
```

# Identity Operators: 
- **'is'** - `Returns True if both the variables refers to the same object (have same memory Address).`
- **'is not'** - `Returns True if both the variables refers to the different object (have different memory Address)`.

```python
# '==' Comparison Operators (data equality) 10 == 10 [Value Check] 
# 'is' Compares the identity (object equality)

a = [1,2,3]
b = a # deep copy
c = [1,2,3] # independent data

print(a == b) # Data Evaluated
print(a is b) # Data + Address

print(a == c) 
print(a is c)
```

```python
a = [1,2,3]
b = a # deep copy
c = ['a',2,3] # independent data

print(a == b) # True
print(a is b) # True

print(a == c) # False
print(a is c) # False
```

```python
# Deep Copy
a = [1,2,3,4,5,6,7]
b = a # Deep Copy
b.append(9)
print(a == b)
print(a is b)
```

```python
# Address of the Variable a & b
# id Function
print(id(a))
print(id(b))
```

```python
a = [1,2,3,4,5,6,7]
b = a.copy() # Shallow Copy
b.append(9)
print(a == b)
print(a is b)
```

```python
# Shallow Copy will create an independent addresses
print(id(a))
print(id(b))
```

```python
print(nested_list[0]) # 111
print(nested_list[2]) # [11, 22, 33, 44, 55]
print(nested_list[2][-1]) # 55
print(nested_list[-2][-5]) # 'a'
print(nested_list[-4][2]) # 33
# print(nested_list[4][5]) # Error
```

```python
# Indexing
# Slicing [Start:0, Stop:Len(Non-Inclusive), Step:+1(default)]
country_list = ['India','America','Ukraine','New Zealand','Australia','South Africa',
               'Russia','Italy','Spain','Germany','Nepal','Ireland','Iran','Pakistan',
               'Canada','India','America','Germany','China']

print(country_list[0]) # 'India'
print(country_list[5]) # 'SA'
print(country_list[-4]) # 'India'
print(country_list[-9]) # 'Nepal'
print(country_list[14]) # 'Canada'
print(country_list[-17]) # 'Ukraine'
```

```python
# Slicing [Start:0, Stop:Len(Non-Inclusive), Step:+1(default)]
# Sub-List
country_list = ['India','America','Ukraine','New Zealand','Australia','South Africa',
               'Russia','Italy','Spain','Germany','Nepal','Ireland','Iran','Pakistan',
               'Canada','India','America','Germany','China']

print(country_list[::]) # Full List
print(country_list[2:5]) # ['Ukraine','New Zealand','Australia']
print(country_list[9:]) # ['Germany', ...... , 'China']
print(country_list[-5:]) # ['Canada' To 'China'] # Step : +1
print(country_list[-9:2]) # []
print(country_list[-9:2:-1]) # ['Nepal' to 'NZ']
print(country_list[-4:-2]) # ['India','America']
```

```python
country_list = ['India','America','Ukraine','New Zealand','Australia','South Africa',
               'Russia','Italy','Spain','Germany','Nepal','Ireland','Iran','Pakistan',
               'Canada','India','America','Germany','China']

print(country_list[-7:1:2]) # []
print(country_list[-7:1:-2]) # [iran->nepal->spain->russia->aus->ukraine]
print(country_list[7::3]) # [italy >> nepal >> pakista >> america]
print(country_list[-9:-2:-1]) # []
print(country_list[-9:-2:2]) # ['Nepal','Iran','Canada','America']
print(country_list[-17:15:4]) # [ukraine->russia->nepal->canada]
```
